### Import Redis

```python
import redis
```

### Connect to a Redis DB

```python
host = "localhost"
```

```python
port = 6379
```

```python
password = ""
```

```python
r = redis.Redis(host, port, password, charset="utf-8", decode_responses=True)
```

```python
r
```

### Set key:value pairs, get values

```python
r.set("name", "Daenerys")
```

```python
r.get("name")
```

```python
r.set("title", "Khaleesi")
```

```python
r.get("title")
```

```python
r.set("net_worth", 4)
```

### Display keys

```python
r.keys()
```

### Increase/Decrease values

```python
r.incr("net_worth")
```

```python
r.decr("net_worth")
```

### Get/Set Expiration time

```python
r.set("relationship", "love")
```

```python
r.expire("relationship", 100)
```

```python
r.ttl("relationship")
```
