 ## INTRODUCTION TO REDIS

Saves key:value pairs on the RAM, can save on disk if size is important.

Used to save the cache version of a website such as cache_id:HTML code

## Syntax:

### Set/Delete pairs and get values

SET Key "Val"

GET Key 

SET user:123 "Name"

GET user:123


DEL user:123

### Increase/Decrease values

INCR money:122

DECR money:122

### Get/Set expiration time

TTL money
EXPIRE money num_seconds_til_xp

###  Lists

#### Create elt : RPUSH

RPUSH books "Daniel Tammett"

#### Create another list before this one: Left Push

LPUSH books "Conan Doyle"

### Get elts list by range

LRANGE books 0 -1

### List length

LLEN books

#### Pop elt

RPOP books

### Sets

unordered list, therefore every key has to be unique

#### Create elt:

SADD users "My dad"

SADD users "Bruce Wayne"

#### Delete elt:

SREM users "Bruce Wayne"

#### List elts of a set:

SMEMBERS users

#### Check if my dad is a member

SISMEMBER users "My dad"

#### Concatenate two sets:

SADD superheroes "Brenda"

SADD superheroes "Roger Federer"

SUNION users superheroes
